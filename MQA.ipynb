```python
import pandas as pd
```

```python
orders_raw = pd.read_csv('orders.csv')
orders_raw['order_date'] = pd.to_datetime(orders_raw['order_date'])
orders_raw['inter_order_gap'] = orders_raw.groupby('customer_id')['order_date'].diff().dt.days
print("Question 1:", orders_raw['inter_order_gap'].median())
```

```python
products_raw = pd.read_csv('products.csv')
highest_price_cogs_ratio = 0.0
best_segment = None 
for segment in products_raw['segment'].unique():
    segment_products = products_raw[products_raw['segment'] == segment].copy()
    segment_products['price_cogs_ratio'] = (segment_products['price'] - segment_products['cogs']) / segment_products['price']
    avg_ratio_segment = segment_products['price_cogs_ratio'].mean()
    if avg_ratio_segment > highest_price_cogs_ratio:
        highest_price_cogs_ratio = avg_ratio_segment
        best_segment = segment

print("Question 2:", best_segment)
```

```python
returns_raw = pd.read_csv('returns.csv')
joined_data = pd.merge(returns_raw, products_raw, on='product_id')

streetwear_returns = joined_data[joined_data['category'] == 'Streetwear']

streetwear_returns_reason = (
    streetwear_returns
    .groupby('return_reason')['return_quantity']
    .sum()
    .idxmax()
)

print("Question 3:", streetwear_returns_reason)
```

```python
web_traffic_raw = pd.read_csv('web_traffic.csv')
result = web_traffic_raw.groupby('traffic_source')['bounce_rate'].mean().idxmin()
print("Question 4:", result)
```

```python
order_items = pd.read_csv('order_items.csv', low_memory=False)

promo_count = order_items['promo_id'].count()
total_rows = len(order_items)

percentage = (promo_count / total_rows) * 100
print(f"Question 5: {percentage:.2f}%")
```

```python
customers = pd.read_csv('customers.csv')
orders = pd.read_csv('orders.csv')

customers = customers[customers['age_group'].notna()]

order_counts = orders.groupby('customer_id').size().reset_index(name='total_orders')
df_merged = pd.merge(customers, order_counts, on='customer_id', how='left')
df_merged['total_orders'] = df_merged['total_orders'].fillna(0)

age_group_stats = df_merged.groupby('age_group').agg(
    total_orders_sum=('total_orders', 'sum'),
    customer_count=('customer_id', 'count')
)

age_group_stats['avg_orders_per_customer'] = age_group_stats['total_orders_sum'] / age_group_stats['customer_count']
best_age_group = age_group_stats['avg_orders_per_customer'].idxmax()

print("Question 6:", best_age_group)
```

```python
orders = pd.read_csv('orders.csv', low_memory=False)
order_items = pd.read_csv('order_items.csv', low_memory=False)
customers = pd.read_csv('customers.csv')
products = pd.read_csv('products.csv')
geography = pd.read_csv('geography.csv')
returns = pd.read_csv('returns.csv')

master_df = pd.merge(orders, order_items, on = 'order_id', how='left')
master_df = pd.merge(master_df, customers, on = ['customer_id', 'zip'], how = 'left')
master_df = pd.merge(master_df, geography, on = 'zip', how = 'left')
master_df = pd.merge(master_df, returns, on=['order_id', 'product_id'], how='left')

master_df.head(10)
```

```python
master_df.columns
```

```python

# Fill null cho các đơn không có hoàn trả
master_df['discount_amount'] = master_df['discount_amount'].fillna(0)
master_df['refund_amount'] = master_df['refund_amount'].fillna(0)

# Gross revenue = số lượng * đơn giá - giảm giá
master_df['gross_revenue'] = (
    master_df['quantity'] * master_df['unit_price']
    - master_df['discount_amount']
)

# Net revenue = gross revenue - tiền hoàn trả
master_df['net_revenue'] = (
    master_df['gross_revenue']
    - master_df['refund_amount']
)

# Tổng hợp theo vùng
region_summary = (
    master_df
    .groupby('region', as_index=False)
    .agg(
        gross_revenue=('gross_revenue', 'sum'),
        total_refund=('refund_amount', 'sum'),
        net_revenue=('net_revenue', 'sum')
    )
    .sort_values('net_revenue', ascending=False)
)

# In toàn bộ để so sánh (nếu có đúng 3 vùng thì sẽ ra đủ 3 vùng)
print("=== Revenue by Region ===")
print(region_summary.to_string(index=False))

# In top 3 vùng có doanh thu cao nhất (phòng trường hợp > 3 vùng)
print("\n=== Top 3 Regions (Net Revenue) ===")
print(region_summary.head(3).to_string(index=False))

# Chênh lệch % giữa top 3 để kiểm tra "xấp xỉ nhau"
top3 = region_summary.head(3).copy()
top3['pct_vs_top1'] = (
    top3['net_revenue'] / top3.iloc[0]['net_revenue'] * 100
)

print("\n=== Relative to Top 1 (%) ===")
print(top3[['region', 'net_revenue', 'pct_vs_top1']].to_string(index=False))
```

```python
cancelled_orders = orders_raw[orders_raw['order_status'] == 'cancelled']
result = cancelled_orders['payment_method'].value_counts().idxmax()

print(f"Question 8: {result}")
```

```python
# Tổng số order_items theo size
order_items_with_size = pd.merge(
    order_items,
    products[['product_id', 'size']],
    on='product_id',
    how='left'
)

order_count = (
    order_items_with_size
    .groupby('size')
    .size()
    .reset_index(name='total_order_items')
)

# Tổng số returns theo size
returns_with_size = pd.merge(
    returns,
    products[['product_id', 'size']],
    on='product_id',
    how='left'
)

return_count = (
    returns_with_size
    .groupby('size')
    .size()
    .reset_index(name='total_returns')
)

# Tính return rate
size_return_rate = pd.merge(
    order_count,
    return_count,
    on='size',
    how='left'
)

size_return_rate['total_returns'] = size_return_rate['total_returns'].fillna(0)

size_return_rate['return_rate'] = (
    size_return_rate['total_returns']
    / size_return_rate['total_order_items']
)

result = size_return_rate.sort_values(
    'return_rate',
    ascending=False
).iloc[0]['size']

print(f"Question 9: {result}")
```

```python
payments = pd.read_csv('payments.csv')

# Giá trị thanh toán trung bình theo số kỳ trả góp
avg_payment_by_plan = (
    payments
    .groupby('installments', as_index=False)['payment_value']
    .mean()
    .sort_values('payment_value', ascending=False)
)

print("=== Average Payment by Installment Plan ===")
print(avg_payment_by_plan.to_string(index=False))

# Kế hoạch có giá trị thanh toán trung bình cao nhất
best_plan = avg_payment_by_plan.iloc[0]['installments']

print(f"\nQuestion 10: {best_plan} kỳ")
```
